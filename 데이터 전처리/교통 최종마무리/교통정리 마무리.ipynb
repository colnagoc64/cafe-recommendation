```python
import pandas as pd
```

## 지하철 전처리

```python
sw= pd.read_csv('지하철 전처리/지하철 승하차인원 평균 월,시간,좌표.csv',encoding='cp949')
```

```python
sw
```

```python
# 버스와 통일시키기위해 필요없는 칼럼 삭제
sw=sw.drop(sw.columns[:3],axis=1)
sw
```

```python
sw.name.value_counts()
```

```python
# groupby를 통해 정거장기준으로 중복된값 합치고 평균치 내기
sw_gp=sw.groupby(['name'],as_index=False).mean()
```

```python
sw_gp
```

```python
# 버스와 칼럼 동일하게하기위해 이름 바꿔줌
sw_gp.rename(columns = {'name' : 'station'}, inplace = True)
sw_gp
```

```python
# 버스에있는 코드를 지하철에서도 임의로 만들어주기위해 버스에없는 숫자를 임의로 부여
# 데이터 합치기위해 임의로 만들어주는코드
# 기존 지하철 코드를 가져와서 사용할시 버스와 겹칠수있기떄문에 상대적으로 작은 값의 지하철에 만듬
code=[]
for i in range(30000,30509):
    code.append(i)
sw_gp['code']=code
sw_gp
```

```python
# 필요없는 소수점값 버리기
sw_gp=sw_gp.astype({
                          '07-10 on avg':'int',
                          '07-10 off avg':'int',
                          '10-17 on avg':'int',
                          '10-17 off avg':'int',
                          '17-22 on avg':'int',
                          '17-22 off avg':'int'})
```

```python
# 칼럼 재정렬
sw_gp=sw_gp[['code','station','07-10 on avg','07-10 off avg','10-17 on avg','10-17 off avg','17-22 on avg','17-22 off avg','X','Y']]
sw_gp
```

## 버스 전처리

```python
bus=pd.read_csv('버스 전처리/버스 전처리.csv',encoding='cp949')
```

```python
bus
```

```python
# groupby 를 통해 중복된값을 합치고 평균치 낸다.
bus1=bus.groupby(['code'],as_index=False).mean()
```

```python
# 이과정에서 이름값은 str 이라 값을 일어버림 그래서 역을붙여줘야함
bus1
```

### 중복값 모두 제거 

```python
# 없어진 역이름을 붙이기위해 역값만 뽑는다.
# station 값에 code가 중복으로 적용되어있음
# 예를들어 숭례문이면 1234 : 숭례문 , 1236:숭례문 버스의 상행 하행이있기때문에 똑같은 정류장이름이있음
# 그래서 유니크값인 코드값을 기준으로 중복된값만 버리고 유니크한 코드갑 1개만 남김 
t=bus.drop_duplicates(['code'])
```

```python
t
```

```python
#필요없는 칼럼 지우고
t=t.drop(t.columns[4:],axis=1)
t=t.drop(t.columns[:2],axis=1)
t
```

```python
#위에 bus1에 정렬한 기준에 맞게 인덱스도 정렬해줌
t=t.sort_values(by=t.columns[0],ascending=True)
t
```

```python
#머지로 두데이터 합쳐서 코드값기준으로 역이름을 붙여준다.
merge_df=pd.merge(bus1,t,how='outer', on='code')
merge_df
```

```python
# 불필요한 칼럼삭제
merge_df=merge_df.drop(merge_df.columns[1:3],axis=1)
merge_df
```

```python
# 불필요한 소수점값없애기
merge_df=merge_df.astype({
                          '07-10 on avg':'int',
                          '07-10 off avg':'int',
                          '10-17 on avg':'int',
                          '10-17 off avg':'int',
                          '17-22 on avg':'int',
                          '17-22 off avg':'int'})
```

```python
merge_df
```

```python
#칼럼값 재설정 위에 지하철과 컨켓시키기 위해
merge_df=merge_df[['code','station','07-10 on avg','07-10 off avg','10-17 on avg','10-17 off avg','17-22 on avg','17-22 off avg','X','Y']]
merge_df
```

## 데이터 합치기
- concat 사용

```python
# 서로다른두데이터 동일한 칼럼값을가지고
# 새로운 칼럼늘 늘리지않고 똑같이 데이터값만 row로 늘리기위해 concat 사용
concat_df=pd.concat([merge_df,sw_gp])
concat_df
```

```python
concat_df.code.value_counts()
```

```python
# 인덱스 재조정
concat_df.reset_index(drop=True, inplace=True)
concat_df
```

## 합친데이터 저장

```python
concat_df.to_csv("교통 정보 최종.csv",encoding='cp949')
```
