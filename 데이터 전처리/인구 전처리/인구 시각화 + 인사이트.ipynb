```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop=pd.read_csv('인구데이터 전처리.csv',encoding='cp949')
```

```python
plt.rcParams['font.family']='Malgun Gothic'
pop.plot.bar(x='gu',alpha=0.5,figsize=(30,10))
plt.title('유동인구',fontsize=15)
plt.ylabel('나이/성별 유동인구',fontsize=10)
plt.xticks(fontsize=12,rotation=0)
plt.legend(loc='upper left')
plt.show()
```

### 인사이트
